```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
import os

from sklearn import tree
import statistics
import scipy.stats as st

import random
import nltk
import glob
nltk.download('punkt')
```

```python
train_sentences = []
val_sentences = []
for filename in sorted(glob.glob(os.path.join('Data', '*txt'))):
    print(filename)
    f = open(filename, 'r', encoding="utf-8")
    file_input = f.readlines()
    for cnt, line in enumerate(file_input):
        sentences = nltk.sent_tokenize(line)
        # print(sentences)
        for i , sent in enumerate(sentences):
            if i <= len(sentences)*0.85:
                train_sentences.append(sent)
            else:
                val_sentences.append(sent)
```

```python
display(train_sentences)
print(len(train_sentences))
```

```python
train = train_sentences[:26000]
dev = train_sentences[26000:30000]
test = train_sentences[30000:]
with open('train.txt', 'w') as f:
    for line in train:
        f.write(line)
        f.write('\n')
with open('dev.txt', 'w') as f:
    for line in dev:
        f.write(line)
        f.write('\n')
with open('test.txt', 'w') as f:
    for line in test:
        f.write(line)
        f.write('\n')
with open('train.label', 'w') as f:
    for line in train:
        f.write("Trump_Speech")
        f.write('\n')
with open('dev.label', 'w') as f:
    for line in dev:
        f.write("Trump_Speech")
        f.write('\n')
with open('test.label', 'w') as f:
    for line in test:
        f.write("Trump_Speech")
        f.write('\n')
        
```
